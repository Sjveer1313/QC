```
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
from qiskit.visualization import plot_histogram
from IPython.display import display, Markdown
import matplotlib.pyplot as plt

# Define circuits here, as they are used in this cell and others later.
qc_a = QuantumCircuit(2)
qc_a.h(1)
qc_a.cz(0, 1)
qc_a.h(1)
qc_a.measure_all()

qc_b = QuantumCircuit(2)
qc_b.h(0)
qc_b.h(1)
qc_b.cx(0, 1)
qc_b.h(0)
qc_b.h(1)
qc_b.measure_all()

qc_c = QuantumCircuit(3)
qc_c.cx(0, 2)
qc_c.cx(1, 2)
qc_c.cx(0, 1)
qc_c.measure_all()

qc_d = QuantumCircuit(3)
qc_d.cx(0, 1)
qc_d.cx(1, 2)
qc_d.cx(0, 1)
qc_d.cx(1, 2)
qc_d.measure_all()


circuits = {
    "Circuit (a): H-CZ-H Identity": qc_a,
    "Circuit (b): CNOT Direction Flip": qc_b,
    "Circuit (c): Multi-CX Logic": qc_c,
    "Circuit (d): CX Chain": qc_d
}

sampler = StatevectorSampler()

for name, qc in circuits.items():
    # 1. Run the simulation
    job = sampler.run([qc])
    result = job.result()[0]
    counts = result.data.meas.get_counts()

    # 2. Setup the figure for side-by-side display
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    # 3. Draw the Circuit
    display(Markdown(f"## {name}"))
    qc.draw(output='mpl', ax=ax1)
    ax1.set_title("Circuit Architecture", fontsize=14, pad=15)

    # 4. Plot the Histogram
    plot_histogram(counts, ax=ax2)
    ax2.set_title("Measurement Statistics", fontsize=14, pad=15)

    plt.tight_layout()
    plt.show()
    display(Markdown("---")) # Divider for clarity
```

```
import math
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
from IPython.display import display, Markdown
import matplotlib.pyplot as plt

gates = {
    "Y Gate": lambda c: c.y(0),
    "Z Gate": lambda c: c.z(0),
    "Phase Gate (with angle pi/4)": lambda c: c.p(math.pi/4, 0),
    "S Gate": lambda c: c.s(0),
    "S-dagger Gate": lambda c: c.sdg(0),
    "T Gate": lambda c: c.t(0),
    "T-dagger Gate": lambda c: c.tdg(0)
}

for gate_name, apply_gate in gates.items():
    display(Markdown(f"**{gate_name}**"))

    qc = QuantumCircuit(1)

    qc.h(0)

    apply_gate(qc)

    # Adjust size of circuit drawing
    fig_circuit = qc.draw(output='mpl')
    fig_circuit.set_size_inches(4, 2)
    display(fig_circuit)

    state = Statevector.from_instruction(qc)

    display(state.draw('latex'))

    # Adjust size of Bloch sphere plot
    bloch_fig = plot_bloch_multivector(state)
    bloch_fig.set_size_inches(3, 3)
    display(bloch_fig)
```
