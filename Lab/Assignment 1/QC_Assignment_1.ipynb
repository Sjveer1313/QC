### Section 2

```python
## •
n = int(input("Enter number of elements : "))
vec = np.array([int(input()) for _ in range(n)])
print(vec)
```

```python
## •
def is_normalised(v):
  return np.isclose(np.linalg.norm(v),1)

print("Is Normalized:", is_normalised(vec))
```

```python
## •
def Product(A,B):
  dot = np.dot(A,B)
  cross = np.cross(A,B)
  return dot,cross

x = np.array([1, 2, 3])
y = np.array([4,5,6])
dot,cross = Product(x,y)
print(dot)
print(cross)
```

```python
## •
def unit_vector(a):
  return a/np.linalg.norm(a)

print(unit_vector(vec))
```

### Section 3

```python
## •
rows = int(input("Enter the no. of rows : "))
cols = int(input("Enter the no. of cols : "))
mat = np.array([[int(input()) for _ in range(cols)] for _ in range(rows)])
print(mat)
```

```python
## •
print(mat.T)
```

```python
if(rows == cols):
  ## •
  print("Determinant : ",np.linalg.det(mat))

  ## •
  print("Rank : ",np.linalg.matrix_rank(mat))

  ## •
  print("Trace : ",np.trace(mat))

  ## •
  eigenvalues,eigenvector = np.linalg.eig(mat)
  print("Eigenvalues : ",eigenvalues)
  print("Eigenvector : \n",eigenvector)

  ## •
  if(np.linalg.det(mat) != 0):
    print("Inverse : \n",np.linalg.inv(mat))
  else:
    print("Inverse doesn't exists")
else:
  print("Matrix must be square for determinant, eigenvalues, and inverse")
```
